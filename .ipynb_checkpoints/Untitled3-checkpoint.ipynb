```python
pk = np.zeros(len(k))
c.calc_P(
            k, 0,
            alpha_perp=1, alpha_parallel=1, b1=2, b2=0, bG2=0, b3=0, bG3=0, bG2d=0, bGamma3=0,
            c0=0, c1=0, c2=0, knl=.3,
            integrand='tree',
            ks=.05
        )
plt.plot(c.PK['kbin'],c.PK['K'],label='tree')
pk += c.PK['K']
c.calc_P(
            k, 0,
            alpha_perp=1, alpha_parallel=1, b1=2, b2=0, bG2=0, b3=0, bG3=0, bG2d=0, bGamma3=0,
            c0=1, c1=0, c2=0, knl=.3,
            integrand='counterterm',
            ks=.05
        )
plt.plot(c.PK['kbin'],c.PK['K'],label='counterterm')
pk += c.PK['K']
c.calc_P(
            k, 0,
            alpha_perp=1, alpha_parallel=1, b1=2, b2=-1, bG2=0.1, b3=0, bG3=0, bG2d=0, bGamma3=-0.1,
            c0=0, c1=0, c2=0, knl=.3,
            integrand='1loop',
            ks=.05
        )
plt.plot(c.PK['kbin'],c.PK['K'],label='1loop')
#plt.plot(c.PK['kbin'],-c.PK['K'],label='1loop-neg')
pk += c.PK['K']
plt.plot(c.PK['kbin'],pk,label='sum')
plt.legend()
#plt.yscale('log')
#plt.xscale('log')
plt.show()
```

```python
b_tree = ['b1_b1','b1_f','f_f']
b_ct = ["c0_b1","c0_f","c1_b1","c1_f","c2_b1","c2_f"]
b_1loop_22 = ["b2_b2",'b2_b2_norm',"b2_bG2","b1_b2","b2_f","b1_b2_f","b2_f_f","bG2_bG2","b1_bG2","bG2_f",
              "b1_bG2_f","bG2_f_f","b1_b1","b1_f","b1_b1_f","b1_f_f","f_f","f_f_f","b1_b1_f_f",
              "b1_f_f_f","f_f_f_f"]
#b_1loop_13 = ["b1_b3","b1_bG3","b1_bG2d","b1_bGamma3","b1_b1","b1_f","b1_b1_f_f","b1_f_f_f",
#              "b1_b1_f","b1_f_f","b1_b2","b1_bG2","b1_b2_f","b1_bG2_f","b3_f","bG3_f","bG2d_f",
#              "bGamma3_f","f_f","f_f_f_f","f_f_f","b2_f","bG2_f","b2_f_f","bG2_f_f"]
b_1loop_13 = ["b1_bGamma3","b1_b1","b1_f","b1_b1_f_f","b1_f_f_f",
              "b1_b1_f","b1_f_f","b1_bG2",
              "bGamma3_f","f_f","f_f_f_f","f_f_f","bG2_f"]

for b in b_tree:
    c.kernel_computation_Pk(b, k, 0, integrand='tree')
    print(b)
    plt.plot(c.PK['kbin'],c.PK['K'])
    plt.plot(c.PK['kbin'],-c.PK['K'],linestyle=':')
    plt.xlim(1e-3,1)
    plt.ylim(1,1e4)
    plt.xscale('log')
    plt.yscale('log')
    plt.show()
for b in b_ct:
    c.kernel_computation_Pk(b, k, 0, integrand='counterterm')
    print(b)
    plt.plot(c.PK['kbin'],c.PK['K'])
    plt.plot(c.PK['kbin'],-c.PK['K'],linestyle=':')
    plt.xlim(1e-3,1)
    plt.ylim(1,1e4)
    plt.xscale('log')
    plt.yscale('log')
    plt.show()
pk = np.zeros(len(k))
for b in b_1loop_22:
    c.kernel_computation_Pk(b, k, 0, integrand='1loop_22')
    print(b)
    tmppk = c.PK['K']
    if b=='b2_b2':
        c.kernel_computation_Pk('b2_b2_norm', k, 0, integrand='1loop_22')
    c.PK['K'] += tmppk
    plt.plot(c.PK['kbin'],c.PK['K'])
    plt.plot(c.PK['kbin'],-c.PK['K'],linestyle=':')
    plt.xlim(1e-3,1)
    plt.ylim(1,1e4)
    plt.xscale('log')
    plt.yscale('log')
    plt.show()
for b in b_1loop_13:
    c.kernel_computation_Pk(b, k, 0, integrand='1loop_13')
    print(b)
    plt.plot(c.PK['kbin'],c.PK['K'])
    plt.plot(c.PK['kbin'],-c.PK['K'],linestyle=':')
    plt.xlim(1e-3,1)
    plt.ylim(1,1e4)
    plt.xscale('log')
    plt.yscale('log')
    plt.show()
plt.xlim(1e-3,1)
plt.ylim(1,1e4)
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
c.calc_P(
            k, 0,
            alpha_perp=1, alpha_parallel=1, b1=2, b2=0, bG2=0, b3=0, bG3=0, bG2d=0, bGamma3=0,
            c0=0, c1=0, c2=0, knl=.3,
            integrand='tree',
            ks=.05
        )
plt.plot(c.PK['kbin'],c.PK['K'],label='tree')
pk_kern = np.zeros_like(c.PK['kbin'])
for b in b_tree:
    bias = 1
    c.kernel_computation_Pk(b, k, 0, integrand='tree')
    if 'b1' in b: bias *= 2**b.count('b1')
    pk_kern += c.PK['K']*bias
plt.plot(c.PK['kbin'],pk_kern)
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
c.calc_P(
            k, 0,
            alpha_perp=1, alpha_parallel=1, b1=2, b2=0, bG2=0, b3=0, bG3=0, bG2d=0, bGamma3=0,
            c0=2, c1=1, c2=-1, knl=.3,
            integrand='counterterm',
            ks=.05
        )
plt.plot(c.PK['kbin'],c.PK['K'],label='tree')
pk_kern = np.zeros_like(c.PK['kbin'])
for b in b_ct:
    bias = 1
    c.kernel_computation_Pk(b, k, 0, integrand='counterterm')
    if 'b1' in b: bias *= 2
    if 'c0' in b: bias *= 2
    if 'c1' in b: bias *= 1
    if 'c2' in b: bias *= -1
    pk_kern += c.PK['K']*bias
plt.plot(c.PK['kbin'],pk_kern)
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
b_tree = ['b1_b1','b1_f','f_f']
b_ct = ["c0_b1","c0_f","c1_b1","c1_f","c2_b1","c2_f"]
b_1loop_22 = ["b2_b2",'b2_b2_norm',"b2_bG2","b1_b2","b2_f","b1_b2_f","b2_f_f","bG2_bG2","b1_bG2","bG2_f",
              "b1_bG2_f","bG2_f_f","b1_b1","b1_f","b1_b1_f","b1_f_f","f_f","f_f_f","b1_b1_f_f",
              "b1_f_f_f","f_f_f_f"]
#b_1loop_13 = ["b1_b3","b1_bG3","b1_bG2d","b1_bGamma3","b1_b1","b1_f","b1_b1_f_f","b1_f_f_f",
#              "b1_b1_f","b1_f_f","b1_b2","b1_bG2","b1_b2_f","b1_bG2_f","b3_f","bG3_f","bG2d_f",
#              "bGamma3_f","f_f","f_f_f_f","f_f_f","b2_f","bG2_f","b2_f_f","bG2_f_f"]
b_1loop_13 = ["b1_bGamma3","b1_b1","b1_f","b1_b1_f_f","b1_f_f_f",
              "b1_b1_f","b1_f_f","b1_bG2",
              "bGamma3_f","f_f","f_f_f_f","f_f_f","bG2_f"]
c.calc_P(
            k, 0,
            alpha_perp=1, alpha_parallel=1, b1=2, b2=-1, bG2=0.1, b3=0, bG3=0, bG2d=0, bGamma3=-.1,
            c0=0, c1=0, c2=0, knl=.3,
            integrand='1loop',
            ks=.05
        )
plt.plot(c.PK['kbin'],c.PK['K'],label='1loop')
pk_kern = np.zeros_like(c.PK['kbin'])
for b in b_1loop_22:
    bias = 1
    c.kernel_computation_Pk(b, k, 0, integrand='1loop_22')
    if 'b1' in b: bias *= 2 #**b.count('b1')
    if 'b2' in b: bias *= -1 #**b.count('b2')
    if 'bG2' in b: bias *= .1 #**b.count('bG2')
    pk_kern += c.PK['K']*bias
for b in b_1loop_13:
    bias = 1
    c.kernel_computation_Pk(b, k, 0, integrand='1loop_13')
    if 'b1' in b: bias *= 2 #**b.count('b1')
    if 'b2' in b: bias *= -1
    if 'bG2' in b: bias *= .1
    if 'bGamma3' in b: bias *= -.1
    pk_kern += c.PK['K']*bias
plt.plot(c.PK['kbin'],pk_kern)
plt.xscale('log')
plt.yscale('log')
plt.show()
```
